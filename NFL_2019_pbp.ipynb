```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read in data and set view options
pd.set_option('display.max_columns', None)
csv_file = 'https://raw.githubusercontent.com/tmbern/NFL_2019/master/Data/pbp-2019.csv'
df = pd.read_csv(csv_file)
print(df.shape)
df.head()
```

```python
# what are the data types?
df.dtypes
```

```python
# sort the values by the GameID
df.sort_values(['GameId', 'Quarter'])
```

```python
df[(df['GameId'] == 2019090500) & (df['Quarter'] == 1)].sort_values(by=['Quarter', 'Minute', 'Second'], ascending=[True, False, False]).head()
```

```python
# get all the games where the lions played

df['OffenseTeam'].unique()
```

```python
# get a dataframe for the lions
lions = df[(df['OffenseTeam'] == 'DET')].sort_values(by=['GameId', 'Quarter', 'Minute', 'Second'], ascending=[True, True, False, False])
```

```python
# what was the distribution of play type on first downs for the lions
lions[lions['Down'] == 1]['PlayType'].value_counts()
```

```python
plt.bar(x=lions['Down'] == 1, height=lions[(lions['PlayType'] == 'PASS')]['PlayType'].count())
```

```python
plt.bar(x = lions[(lions['Down'] == 1) & (lions['PlayType'] == 'PASS')]['Down'], height=lions[(lions['Down'] == 1) & (lions['PlayType'] == 'PASS')]['PlayType'].count(), width=1)
```
